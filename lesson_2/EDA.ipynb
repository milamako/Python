```python
from data_loader import DataLoader
from data_cleaner import DataCleaner
from visualization import VisualizationModule
```

Загрузка и чтение файла

```python
loader = DataLoader('pokemon.csv')
data = loader.load_data()
print(data)
```

```python
viz_module = VisualizationModule(data)

# Установка цветовой схемы
viz_module.set_color_palette('viridis')

# Визуализация категориальных данных
viz_module.bar_chart('type1')
viz_module.pie_chart('type1')
viz_module.donut_chart ('type1')
# Визуализация числовых данных
viz_module.histogram('attack')
viz_module.box_plot('attack')
# Диаграмма рассеяния и линейный график
viz_module.scatter_plot('attack', 'defense')
viz_module.line_chart('generation', 'base_total')

# Тепловая карта
viz_module.heatmap()
```

Очистка графиков

```python
viz_module.clear_plots()
```

Подсчёт и вывод пропущенных значений, вывод категориальных и числовых столбцов.

```python

cleaner = DataCleaner(data)
cleaner.report_missing_values()
```

```python
cleaner.report_column_types()
```

Заполняет пропущенные значения в числовых и категориальных столбцах.
    Для числовых данных можно выбрать стратегию: 'mean', 'median', 'mode'.
    Категориальные данные всегда заполняются наиболее частым значением ('mode').

```python
cleaner.fill_missing_values(strategy='median')
```
